```python
from sklearn import datasets

iris_data = datasets.load_iris()
```

```python
x = iris_data['data']
y = iris_data['target']
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold

model = DecisionTreeClassifier(max_depth=2)
# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Create a 10-fold cross-validation object
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Evaluate the decision tree using 10-fold cross-validation
scores = cross_val_score(model, x, y, cv=kf)

# Print the average score and standard deviation
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import cross_val_score, LeaveOneOut

# Load the iris dataset
model = DecisionTreeClassifier(max_depth=2)

# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Create a Leave-one-out cross-validation object
loo = LeaveOneOut()

# Evaluate the decision tree using Leave-one-out cross-validation
scores = cross_val_score(model, x , y , cv=loo)

# Print the average score and standard deviation
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
